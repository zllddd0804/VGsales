```python
# 处理缺失值
```

```python
df.isnull().sum().sort_values(ascending=False)    #查看缺失值情况
```

```python
# 计算缺失值比例
df_null = df.isnull().sum()/len(df)

# 转变为百分数
df_null.apply(lambda x: format(x, '.1%')) 
```

```python
for item in ['发售时间']:
    df[item] = df[item].fillna(2021)
# 没有发售时间的统计归为2021
for item in ['发行公司']:
    df[item] = df[item].fillna(0)
```

```python
# 以下为逐年各地区销售额情况
```

```python
df_NA=df['NA_Sales'].groupby(df['发售时间']).sum().reset_index()
df_EU=df['EU_Sales'].groupby(df['发售时间']).sum().reset_index()
df_JP=df['JP_Sales'].groupby(df['发售时间']).sum().reset_index()
df_OT=df['Other_Sales'].groupby(df['发售时间']).sum().reset_index()
df_GL=df['Global_Sales'].groupby(df['发售时间']).sum().reset_index()
# 将全国每个地区的每年销售额做成表
# 另一个做法df_NA=df['NA_Sales'].groupby(df['发售时间']).apply(lambda x: x.sum())
```

```python
na_x=df_NA.iloc[:,0].values
na_y=df_NA.iloc[:,1].values
eu_x=df_EU.iloc[:,0].values
eu_y=df_EU.iloc[:,1].values
jp_x=df_JP.iloc[:,0].values
jp_y=df_JP.iloc[:,1].values
ot_x=df_OT.iloc[:,0].values
ot_y=df_OT.iloc[:,1].values
# 为实现pyecharts折线图数据准备
```

```python
from matplotlib.pyplot import MultipleLocator
plt.figure(figsize=(15,8))#设置画布的尺寸
ax=plt.gca() #ax为两条坐标轴的实例
plt.title('全球各地区销售额数据折现图',fontsize=20)#标题，并设定字号大小
plt.xlabel('年份',fontsize=14)#设置x轴，并设定字号大小
plt.ylabel('销售额',fontsize=14)#设置y轴，并设定字号大小
x_major_locator=MultipleLocator(2)#把x轴的刻度间隔设置为2，并存在变量里
ax.xaxis.set_major_locator(x_major_locator)
#把x轴的主刻度设置为2的倍数

plt.plot(df_NA['发售时间'],df_NA['NA_Sales'],color='darkblue',label='NA_Sales',marker='p')
plt.plot(df_EU['发售时间'],df_EU['EU_Sales'],color='darkseagreen',label='EU_Sales',marker='+')
plt.plot(df_JP['发售时间'],df_JP['JP_Sales'],color='deeppink',label='JP_Sales',marker='s')
plt.plot(df_GL['发售时间'],df_GL['Global_Sales'],color='gold',label='Global_Sales',marker='d')
plt.plot(df_OT['发售时间'],df_OT['Other_Sales'],color='red',label='Other_Sales',marker='^')
# 将每个地区的变化表画入图中

plt.legend(loc=2)#图例展示位置，数字代表第几象限
plt.show()#显示图像
```

```python
df_NA=df['NA_Sales'].groupby(df['类型']).sum().reset_index()
df_EU=df['EU_Sales'].groupby(df['类型']).sum().reset_index()
df_JP=df['JP_Sales'].groupby(df['类型']).sum().reset_index()
df_OT=df['Other_Sales'].groupby(df['类型']).sum().reset_index()
df_GL=df['Global_Sales'].groupby(df['类型']).sum().reset_index()
# 将全国每个地区的每年销售额做成表
df_GL
```

```python
gl_x=df_GL.iloc[:,0].values
gl_y=df_GL.iloc[:,1].values
```

```python
# 以下显示全球最热门类型Action和最冷门类型Strategy逐年销售额折线图对比
```

```python
df_Type=df['Global_Sales'].groupby([df['类型'],df['发售时间']]).sum().reset_index()
df_Action=df_Type[df_Type['类型']=='Action']
df_Strategy=df_Type[df_Type['类型']=='Strategy']
```

```python
plt.figure(figsize=(15,8))#设置画布的尺寸
ax=plt.gca() #ax为两条坐标轴的实例
plt.title('全球最热门类型Action和最冷门类型Strategy逐年销售额折线图对比',fontsize=20)#标题，并设定字号大小
plt.xlabel('发售时间',fontsize=14)#设置x轴，并设定字号大小
plt.ylabel('销售额',fontsize=14)#设置y轴，并设定字号大小
x_major_locator=MultipleLocator(2)#把x轴的刻度间隔设置为2，并存在变量里
ax.xaxis.set_major_locator(x_major_locator)
#把x轴的主刻度设置为2的倍数

plt.plot(df_Action['发售时间'],df_Action['Global_Sales'],color='darkblue',label='Action',marker='p')
plt.plot(df_Strategy['发售时间'],df_Strategy['Global_Sales'],color='pink',label='Strategy',marker='*')

plt.legend(loc=2)#图例展示位置，数字代表第几象限
plt.show()
```

```python
# 94年和95年冷门类型strategy竟然超过了Action
```

```python
# 查看有多少个平台
df['平台'].unique()
```

```python
# 各平台各游戏类型个数
df_PT=df['类型'].groupby([df['平台'],df['类型']]).count()
df_PT
```

```python
# unstack把index转成columns
df_PT.unstack()
```

```python
# 将count创建为新列
df_PT=df['类型'].groupby([df['平台'],df['类型']]).count().reset_index(name="count")
df_PT
```

```python
# 将每个平台最多的游戏的类型拿出来
df_MAX=df_PT.groupby('平台')['count'].idxmax()
df_MAX=df_PT.iloc[df_MAX]
df_MAX
```

```python
plt.figure(figsize=(20,15))
plt.scatter(df_MAX['平台'],df_MAX['类型'],c=df_MAX['count'],alpha=0.6,cmap="magma",s=150)
plt.xlabel("平台")
plt.ylabel("类型")
plt.title('每个平台数量最多的游戏类型',fontsize=30,color='purple')
plt.colorbar()
plt.show()
```

```python
# 颜色深浅代表最热门游戏数量
```

```python
from pyecharts import Line
line=Line("全球各地区销售额数据折现图",width=1200, height=600)
line.add('北美销量',na_x,na_y)
line.add('欧洲销量',eu_x,eu_y)
line.add('日本销量',jp_x,jp_y)
line.add('其他地区销量',ot_x,ot_y)
line.render('全球各地区销售额数据折现图.html')
```

```python
from pyecharts import Pie
pie=Pie("各类型全球销售额占比",width=1500, height=600)
pie.add('',gl_x,gl_y)
pie.render('各类型全球销售额占比.html') 
```
